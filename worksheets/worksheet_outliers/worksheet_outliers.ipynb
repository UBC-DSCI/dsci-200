### Weighted Average

It can be shown that M-estimators of location can be computed as weighted averages. The weight assigned to each observation depends on how far it is from the center of the data. Observations close to the center receive larger weights, while observations farther away receive smaller weights. Because the weights themselves depend on the center, iterative algorithms are typically used to compute the estimate (details are beyond the scope of this course). In particular, the bisquare M-estimator can assign a weight of 0 to extreme observations. 

> The weight function is derived from the derivative of the $\rho$-function.

Let’s visualize the bisquare weights assigned to each point in `V169`, and overlaying the theoretical bisquare weight function (shown as a dashed red curve).

*Run the code below*

```r
c <- 4.685

u <- answer3$bisquare

# Bisquare weights
w <- ifelse(abs(u) <= c,
            (1 - (abs(u)/c)^2)^2,
            0)

df <- data.frame(u = u, w = w) 

ggplot(df, aes(u, w)) + 
    geom_point(size = 2) + 
    stat_function( fun = function(x) 
        ifelse(abs(x) <= c, (1 - (abs(x)/c)^2)^2, 0), 
        color = "red", 
        xlim = c(-9, 9), 
        linetype = "dashed", linewidth = 0.6 ) + 
        labs( x = "Standardized distance", y = "Bisquare weight" ) + 
        theme_minimal()
```

**Rowwise methods** detect entire observations that are unusual relative to the rest of the dataset, typically using a multivariate distance measure.

**Cellwise methods** instead identify individual cells that are inconsistent with the rest of the data.

### `cellMap`

The function `cellMap` can be used to visualize the results of cellwise outlier detection. It displays a map of the data matrix where flagged cells and rows can be identified and interpreted. The function cellMap displays the data matrix in blocks of rows and columns (5 by default).

*Run the code below to produce a visualization to flag multivariate outliers.*

```r
# plot settings
n <- nrow(glass_clean)
d <- ncol(glass_clean)

nrowsinblock <- 5
ncolumnsinblock <- 5

# rowlabels must have length n
rowlabels <- rep("", n)
rowlabels[1] <- "1"
rowlabels[n] <- "n"

# columnlabels must have length d
columnlabels <- rep("", d)
columnlabels[1] <- "1"
columnlabels[d] <- "d"

rowtitle <- "glass samples"
columntitle <- "wavelengths"

ggpfastDDC <- cellMap(
  D = glass_clean,
  R = fastDDCglass$stdResid,
  indcells = fastDDCglass$indcells,
  indrows = fastDDCglass$indrows,
  rowlabels = rowlabels,
  columnlabels = columnlabels,
  mTitle = "Cellwise Outlier Detection",
  nrowsinblock = nrowsinblock,
  ncolumnsinblock = ncolumnsinblock,
  columnangle = 0,
  rowtitle = rowtitle,
  columntitle = columntitle,
  sizetitles = 1.5,
  autolabel = FALSE
)

plot(ggpfastDDC)
```

Robust statistics is a large and active field that develops methods designed to perform well even when data contain outliers or deviate from ideal assumptions.

In this worksheet we saw that:

- Classical estimators can be strongly affected by extreme observations.
