## Data Privacy

Data privacy entails protecting individuals within a dataset from harm by preventing unauthorized access, use, or re-identification of their personal information. Even when names are removed, data can still be linked back to people (a phenomenon known as the **Mosaic effect**).  

We will be working with a synthetic dataset about a list of job applicants for a fictional tech startup based in Vancouver (https://ubc-dsci.github.io/dsci-200/schedule/data/web-data.html).
